```python
# importing libraries
from bs4 import BeautifulSoup
import pandas as pd 
import requests 
```

```python
# geting url for the response:
url = "https://webscraper.io/test-sites/e-commerce/allinone"
# if we use the only url to get response it gets error:
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
page = requests.get(url, headers=headers)
```

```python
# page data: 
soup = BeautifulSoup(page.content,"html.parser")
# page html is:
soup
```

```python
# geting the main div for the get data from this:
products = soup.find_all("div",class_="card-body")
# now we have all the data for the csv file:
products
```

```python
# creating lists for all the products:
Image_urls = []
products_title = []
products_prices = []
products_description = []
products_rating = []
# now we create a loop for the data gethering:
for i in products:
    # for products images
    Image_url = i.find("img", class_ = "image").get("src")
    Image_urls.append(Image_url)
    
    # for products title
    title = i.find("a", class_="title").text
    products_title.append(title)
    
    # for products prices:
    price = i.find("h4", class_="price").text
    products_prices.append(price[1:])
    
    # for product decription:
    dec = i.find("p",class_="description").text
    products_description.append(dec)
    
    # for products rating:
    rating = i.find("p", attrs={"data-rating": True}).get("data-rating")
    products_rating.append(rating)
```

```python
# creating a dict for the all data
data = {"image":Image_urls, "title":products_title,"price":products_prices,"description":products_description,"rating":products_rating}
page_products = pd.DataFrame(data)
```
